**Basics of Plotting with Numpy and Matplotlib**

```python
import numpy as np
import matplotlib.pyplot as plt

# define time
t = np.linspace(-5,5)

# define the function
x = t**2

# plot the results
plt.plot(t,x)
plt.show()
```
